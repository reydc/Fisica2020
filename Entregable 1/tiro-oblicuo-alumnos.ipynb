Obtenga las trayectorias de la bala (ùë¶(ùë•)) para distintos √°ngulo de elevaci√≥n entre $0^o$ y $90^o$ y grafique.  
Use ùë£=200ùëö/ùë† para implementar los gr√°ficos.

a) Hallar la ecuaci√≥n de movimiento para las coordenadas $x$ e $y$, usando que:

$v_x(t=0) = v \cos(\theta)$

$v_y(t=0) = v \sin(\theta)$,

d√≥nde $\theta$ es el √°ngulo de elevaci√≥n del ca√±on. 

[[ Alumno ]] Sea $g = 9.8\ [m/s^2]$. Desarrollamos las ecuaciones de movimiento:  

$v_x(t) = v \cos(\theta)$  

$v_y(t) = v \sin(\theta) - g \times t$  

$\therefore$  

$x(t) = v \cos(\theta) \times t$  

$y(t) = v \sin(\theta) \times t - \dfrac{g}{2} \times t^2$  


```python
import numpy as np # linear algebra
from matplotlib import pyplot as plt #Para hacer gr√°ficos

v = 200 # m√≥dulo de la velocidad en m/s
g = 9.8 # m√≥dulo de la aceleraci√≥n de la gravedad en m/s¬≤
tita = 60 * np.pi/180 # usamos este valor fijo (60 grados) solo para este ejemplo. 

print('Creamos un vector t con valores ') # El comando print sirve para imprimir mensajes en la consola
t = np.arange(0., 40., 1., dtype='float') # Esta funci√≥n de numpy crea un array de n√∫meros entre 0 y 40 en pasos de 1. La unidad es s 

#Ejemplo: Gr√°fico de x(t)

x = v*np.cos(tita)*t 

print("Graficamos x(t)")
plt.axis([0,40,0,3000]) # Estable los valores minimos y m√°ximo para cada coordenada para el gr√°fico
plt.plot(t, x, '.b') # Grafica los pares ordenados (t,x) con '.' en color azul
plt.title("x versus t, $\\theta = 60^o$")
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel("t [s]")
plt.ylabel("x [m]")
plt.show()

#Ejemplo: Gr√°fico de y(t)

y = v*np.sin(tita)*t-g/2*t**2

print("Graficamos y(t)")

plt.axis([0,40,0,2000]) # Estable los valores minimos y m√°ximo para cada coordenada para el gr√°fico
plt.plot(t, y, '.g') # Grafica los pares ordenados (t,y) con '.' en color verde
plt.title("y versus t, $\\theta = 60^o$")
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel("t [s]")
plt.ylabel("y [m]")
plt.show()

```

[[ Alumno ]] Los pasos desarrollados ser√≠an:  

* Hallar $t(x)$ y luego reemplazar esta funci√≥n en $y(t)$  
* Verificar anal√≠ticamente: $y= {x \tan(\theta)} - {g \over 2} \left({x \over v \cos(\theta)}\right)^2$  

$x(t) = x = v \cos(\theta) \times t \Longleftrightarrow t(x) = \dfrac{x}{v \cos(\theta)}$  

$\therefore\ \ y(x) = y(t(x)) = v \sin(\theta) \times \dfrac{x}{v \cos(\theta)} - \dfrac{g}{2} \times \left(\dfrac{x}{v \cos(\theta)}\right)^2 = x \tan(\theta) - \dfrac{g}{2} \times \left(\dfrac{x}{v \cos(\theta)}\right)^2$  

* Verificar que el alcance del tiro es $x(y=0) = {v^2 \over g} \sin(2 \theta)$  

$y(t) = 0$ s√≥lo para:  

$
\begin{cases}
t_1 = 0\ [s] \\  
t_2 = \dfrac{2v \sin(\theta)}{g} \\
\end{cases}
$  

Entonces:

$Alcance = x(t_2) = v \cos(\theta) \times \dfrac{2v \sin(\theta)}{g} = \dfrac{2v^2 \cos(\theta) \sin(\theta)}{g} =\dfrac{v^2 \times (2 \cos(\theta) \sin(\theta))}{g} = \dfrac{v^2 \sin(2\theta)}{g} $  

teniendo en cuenta que: 

$2 \cos(\theta) \sin(\theta) = 2 \dfrac{e^{i\theta} + e^{-i\theta}}{2} \dfrac{e^{i\theta} - e^{-i\theta}}{2i} = \dfrac{(e^{i\theta})^2 - (e^{-i\theta})^2}{2i} = \dfrac{e^{i(2\theta)} - e^{-i(2\theta)}}{2i} = \sin(2 \theta)$  


```python
print("Crear un array con valores de \u03B8")
print("Resolver para cada \u03B8 usando un loop")
print("\tPara cada \u03B8 hallar previamente el alcance")
print("\tGraficar las trayectorias para cada \u03B8")
```

```python
import numpy as np
from matplotlib import pyplot as plt

v = 200 # Velocidad inicial
g = 9.8 # Aceleraci√≥n de la gravedad

t = np.arange(0, 100, .5)

tita_grados = np.linspace(0, 90, 18)

def radianes(tita):
    return tita * np.pi/180

def x_segun_t(tita, t):
    return v * np.cos(tita) * t

def y_segun_x(tita, x):
    return x * np.tan(tita) - ( (g/2) * x**2 / (v * np.cos(tita))**2 )

def alcance(tita):
    return v**2 * np.sin(2 * tita) / g

for tita in tita_grados:
    alcance_de_plt = alcance(radianes(tita))

    x_de_plt = x_segun_t(radianes(tita), t)

    x_de_plt = x_de_plt[x_de_plt <= alcance_de_plt]
    
    y_de_plt = y_segun_x(radianes(tita), x_de_plt)

    plt.axis([0,4500,0,3000])
    plt.grid()
    plt.plot(x_de_plt, y_de_plt, '.')
    plt.title("x versus y, $\\theta = {}^o$".format(np.around(tita, 7)))
    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
    plt.xlabel("x [m]")
    plt.ylabel("y [m]")
    plt.show()

```
