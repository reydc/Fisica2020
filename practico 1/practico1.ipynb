### Ejercicio 1
Tenemos la función 
$x(t) = 1\ [m/s^2] \times t^2 - 3\ [m/s] \times t$

(a) Gráfica

```python
import numpy as np
from matplotlib import pyplot as plt

t = np.linspace(-10,10,100)
x = t**2 - 3*t

plt.axis([-10, 10, -5, 10])
plt.grid()
plt.plot(t, x, label="$x(t)$")
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel("t [s]")
plt.ylabel("x [m]")
plt.legend(loc="upper left")
plt.show()
```

(b)  
Para determinar los valores de velocidad media para el móvil usamos que $v_{med} = \frac{ \Delta x }{\Delta t}$  
Graficamente es la pendiente de la secante entre los puntos que nos da la función en los intervalos

```python
from IPython.display import display, Math

def x(t):
  return t**2 - 3 * t

def v_med(ti, tf):
  return (x(tf) - x(ti)) / (tf - ti)

ti = -1.
tf = [5., 4., 1., -0.5, -0.8, -0.9, -0.99, -0.999, -0.9999]

for stop in tf:
  display(Math(r'[{}, {}]  \implies v_{{med}} = {}\ [\frac{{m}}{{s}}]'.format(ti, stop, v_med(ti, stop))))
```

(c)  
Tenemos que $\Delta{t_n} = t_n - t_0$. Entonces vemos que $-1 + \Delta{t_n} - (-1) = -1 + \Delta{t_n} + 1  = \Delta{t_n}$ y $v_{med} = \frac{ x(t_0 + \Delta{t_n}) - x(t_0)}{ \Delta{t_n} }$,  
es decir el cociente de diferencia respecto a $t$.  
A medida que el valor se haga más pequeño, tenderá a la derivada de x respecto a t. El valor al que se aproxima es $5 [\frac{{m}}{{s}}]$.  
Geométricamente lo ocurre es que nos acercamos a la recta tangente en el punto $t_0$.

(d)  
Para encontrar la ecuación de la recta tangente en $t_0$ usamos que la derivada de $v_{med}$ va a ser la pendiente de la recta  
tangente en el punto dado. Por ello evaluamos la derivada en $t_0$:  
$\frac{ dx}{d t}\rvert_{t_0} = 2\ [m/s^2] \times t_0 -3\ [m/s] = -5\ [m/s]$  
Entonces tengo una ecuación de la forma $-5\ [m/s] \times t + c$, pero sabemos que $x(-1\ [s]) = 4 [m]$ por lo tanto $c = -1\ [m]$,  
es decir la ecuación es:  
$-5\ [m/s] \times t - 1\ [m]$


```python
import numpy as np
from matplotlib import pyplot as plt

t = np.linspace(-10,10,100)
x = t**2 - 3 * t
r = -5 * t - 1 

plt.axis([-10, 10, -5, 10])
plt.grid()
plt.plot(t, x, label="$x(t)$")
plt.plot(t, r, label="tangente a $t_0$")
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel("t [s]")
plt.ylabel("x [m]")
plt.legend(loc="upper left")
plt.show()
```
