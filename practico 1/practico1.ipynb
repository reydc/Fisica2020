### Ejercicio 1
Tenemos la función 
$x(t) = 1\ [m/s^2] \times t^2 - 3\ [m/s] \times t$

(a) Gráfica

```python
import numpy as np
from matplotlib import pyplot as plt

t = np.linspace(-10,10,100)
x = t**2 - 3*t

plt.axis([-10, 10, -5, 10])
plt.grid()
plt.plot(t, x, label="$x(t)$")
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel("t [s]")
plt.ylabel("x [m]")
plt.legend(loc="upper left")
plt.show()
```

(b) Para determinar los valores de velocidad media para el móvil usamos que $v_{med} = \frac{ \Delta x }{\Delta t}$  
Graficamente es la pendiente de la secante entre los puntos que nos da la función en los intervalos

```python
from IPython.display import display, Math

def x(t):
  return t**2 - 3 * t

def v_med(ti, tf):
  return (x(tf) - x(ti)) / (tf - ti)

ti = -1.
tf = [5., 4., 1., -0.5, -0.8, -0.9, -0.99, -0.999, -0.9999]

for stop in tf:
  display(Math(r'[{}, {}]  \implies v_{{med}} = {}\ [\frac{{m}}{{s}}]'.format(ti, stop, v_med(ti, stop))))
```

(c) Tenemos que $\Delta{t_n} = t_n - t_0$. Entonces vemos que $-1 + \Delta{t_n} - (-1) = -1 + \Delta{t_n} + 1  = \Delta{t_n}$ y $v_{med} = \frac{ x(t_0 + \Delta{t_n}) - x(t_0)}{ \Delta{t_n} }$,  
es decir el cociente de diferencia respecto a $t$.  
A medida que el valor se haga más pequeño, tenderá a la derivada de x respecto a t. El valor al que se aproxima es $5 [\frac{{m}}{{s}}]$.  
Geométricamente lo ocurre es que nos acercamos a la recta tangente en el punto $t_0$.  

(d) Para encontrar la ecuación de la recta tangente en $t_0$ usamos que la derivada de $v_{med}$ va a ser la pendiente de la recta  
tangente en el punto dado. Por ello evaluamos la derivada en $t_0$:  
$\frac{ dx}{d t}\rvert_{t_0} = 2\ [m/s^2] \times t_0 -3\ [m/s] = -5\ [m/s]$  
Entonces tengo una ecuación de la forma $-5\ [m/s] \times t + c$, pero sabemos que $x(-1\ [s]) = 4 [m]$ por lo tanto $c = -1\ [m]$,  
es decir la ecuación es:  
$-5\ [m/s] \times t - 1\ [m]$  


```python
import numpy as np
from matplotlib import pyplot as plt

t = np.linspace(-10,10,100)
x = t**2 - 3 * t
r = -5 * t - 1 

plt.axis([-10, 10, -5, 10])
plt.grid()
plt.plot(t, x, label="$x(t)$")
plt.plot(t, r, label="tangente a $t_0$")
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel("t [s]")
plt.ylabel("x [m]")
plt.legend(loc="upper left")
plt.show()
```

(d) Ya vimos algunas expresiones para los tres intervalos. Juntamos lo que teniamos para los intervalos  
y hacemos de cuenta que se extiende hacía ambas direcciones de $t$:

$$
\begin{align}
x(t) =
\begin{cases}
-120\ [Km] + 20\ [Km/s]\ t & t \leqslant 6 \\
-120\ [Km] - 28\ [Km/s]\ t + 4\ [Km/s^2]\ t^2 & 6 < t \leqslant 9 \\

(e) Gráficas


```python
import numpy as np
from matplotlib import pyplot as plt

def x_piecewise(t):
  if (t <= 6):
    return -120 + 20 * t
  elif (6 < t and t <= 9):
    return -120 - 28 * t + 4 * t**2
  else:
    return -120 + 110 * t - (11 / 3) * t**2

def v_piecewise(t):
  if (t <= 6):
    return 20
  elif (6 < t and t <= 9):
    return -28 + 8 * t
  else:
    return 110 - (22 / 3) * t

def a_piecewise(t):
  if (t <= 6):
    return 0
  elif (6 < t and t <= 9):
    return 8
  else:
    return - (22 / 3)

t = np.linspace(-20,20,100)

fig, (p1, p2, p3) = plt.subplots(1, 3)
fig.suptitle("Ejercicio 5")
fig.tight_layout(pad = 4)

#p1.axis([-10, 10, -20, 20])
p1.plot(t, list(map(x_piecewise, t)), color="royalblue")
p1.set_title("$x(t)$")
p1.axhline(0, color='black')
p1.axvline(0, color='black')
p1.grid()
p1.set_xlabel("$t\ [s]$")
p1.set_ylabel("$x\ [Km]$")

p2.plot(t, list(map(v_piecewise, t)), color="orange")
p2.set_title("$v(t)$")
p2.axhline(0, color='black')
p2.axvline(0, color='black')
p2.grid()
p2.set_xlabel("$t\ [s]$")
p2.set_ylabel("$v\ [Km/s]$")

p3.plot(t, list(map(a_piecewise, t)), color="red")
p3.set_title("$a(t)$")
p3.axhline(0, color='black')
p3.axvline(0, color='black')
p3.grid()
p3.set_xlabel("$t\ [s]$")
p3.set_ylabel("$a\ [Km/s^2]$")

plt.show()
```

$\vec{a}(t) = - g\ \boldsymbol{\hat{\jmath}}$  
$\vec{v}(t) = (v_0 - g\ t)\ \boldsymbol{\hat{\jmath}}$  
$\vec{y}(t) = (v_0\ t - \frac{g}{2}\ t^2)\ \boldsymbol{\hat{\jmath}}$  

El tiempo de llegada de la pelota al suelo lo obtendríamos anulando la componente vértical:  
$\begin{align}
0 &= v_0\ t - \frac{g}{2}\ t^2\\
\frac{g}{2}\ t^2 &= v_0\ t \\
t &= \frac{2\ v_0}{g}\\
\end{align}
$  
El insante en que se anula la componente vértical de la velocidad es:  
$v_0 - g\ t = 0 \implies t = \frac{v_0}{g}$  
Entonces la altura máxima a la que llega la pelota es:  
$v_0\ \dfrac{v_0}{g} - \dfrac{g}{2}\ (\dfrac{v_0}{g})^2 = v_0^2\ (\dfrac{1}{g} - \dfrac{1}{2\ g}) = \dfrac{v_0^2}{2\ g}$  

La gráfica del desplazamiento vértical nos mostraría una parábola con ramas hacia abajo.  
La gráfica de la velocidad nos mostraría una recta que cruza desde las izquierda superior hacia la derecha inferior.  
La gráfica de la aceleración es constante.

$\vec{y}(t) = (v_0\ t - \frac{g}{2}\ t^2)\ \boldsymbol{\hat{\jmath}}$  

Ahora resulta que para los instantes $t_A$ y $t_B$ se cumple que  

$A = v_0\ t_A - \dfrac{g}{2}\ t_A^2$  
$B = v_0\ t_B - \dfrac{g}{2}\ t_B^2 = A + 3\ [m]$  

y  

$v = v_0 - g\ t_A$  
$v/2 = v_0 - g\ t_B$  

Restando las primeras a las segundas obtenemos

$3\ [m] = v_0\ (t_B - t_A) - \frac{g}{2}\ (t_B^2 - t_A^2)$  

$t_B - t_A = \dfrac{v}{2\ g}$  
$v_0 = v + g\ t_A$

$\therefore$

$
\begin{align}
3\ [m] &= (v + g\ t_A)\ \dfrac{v}{2\ g} - \dfrac{g}{2}\ \dfrac{v}{2\ g}\ (t_B + t_A) \\
3\ [m] &= [v + g\ t_A - \dfrac{g}{2}\ (t_B + t_A)]\ \dfrac{v}{2\ g} \\
3\ [m] &= [v + g\ t_A - \dfrac{g}{2}\ t_B - \dfrac{g}{2}\ t_A]\ \dfrac{v}{2\ g} \\
3\ [m] &= [v - \dfrac{g}{2}\ t_B + \dfrac{g}{2}\ t_A]\ \dfrac{v}{2\ g} \\

3\ [m] &= [v - \dfrac{g}{2}\ \dfrac{v}{2\ g}]\ \dfrac{v}{2\ g} \\
3\ [m] &= [v - \dfrac{v}{4}]\ \dfrac{v}{2\ g} \\
3\ [m] &= \dfrac{3\ v}{4}\ \dfrac{v}{2\ g} \\
3\ [m] &= \dfrac{3\ v^2}{8\ g} \\
\end{align}
$  

$\therefore\ \ v = \sqrt{8\ g}$

$v_A(t) = v - g\ t = \sqrt{8\ g} - g\ t$

El altura máxima es la misma si usamos esta referencia. La velocidad será cero en ese momento:  
$t_{max} = \dfrac{\sqrt{8\ g}}{g}$  
$
\begin{align}
y_{max}(t_{max}) &= A + \sqrt{8\ g}\ \dfrac{\sqrt{8\ g}}{g} - \dfrac{g}{2}\ (\dfrac{\sqrt{8\ g}}{g})^2\\
